## Handling Outliers df['Year_Birth']

```python
## Fungsinya buat melihat pesebaran data pada kolom 'Year Birth'
plt.figure(figsize=(10,8))
plt.boxplot(df['Year_Birth'])
plt.title("Boxplot Year Birth")
plt.ylabel("Year_Birth")
plt.show()

### kolom Year_ Birth ada outliers
```

```python
##cari data outliers pada kolom Year_Birth
Q1=df['Year_Birth'].describe()['25%']
Q3=df['Year_Birth'].describe()['75%']
IQR=Q3-Q1
Upper_Fence = Q3 + (1.5 * IQR)
Lower_Fence = Q1 - (1.5 * IQR)
```

```python
#### Cek Data Outliers
df[(df['Year_Birth'] < Lower_Fence) | (df['Year_Birth'] > Upper_Fence)] # membaca data frame yang mengandung data outliers dari kolom 'Year Birth' 
#(df[(df['Year_Birth'] < Lower_Fence) | (df['Year_Birth'] > Upper_Fence)].count()).loc['Income'] #melihat jumlah outliers
```

##### Dari DataFrame berisi outliers diatas, terdapat 3 data yang outliers pada kolom Year Birth. Karena hal tersebut diduga sebagai kesalahan penulisan maka handling outliers yang kami pilih yaitu DROP

```python
# mengakses dataset selain dataset yang mengandung outliers
df=df[~(df['Year_Birth'] < Lower_Fence) | (df['Year_Birth'] > Upper_Fence)]
```

## Update kolom 'Year_Birth' dari tahun kelahiran menjadi kategori generasi

```python
# Fungsi kategorikal 'Year Birth' berdasarkan tahun kelahiran
# sumber wikipedia
def func(row): 
    if row['Year_Birth'] <= 1945 : #diatas umur 75 tahun
        val = "Silent Generation"
    if row['Year_Birth'] > 1945 and row['Year_Birth'] <= 1965  : # 55-74
        val = "Boomers"
    elif row['Year_Birth'] > 1965 and row['Year_Birth'] <= 1980  : # 40-54
        val = "Gen X"
    elif row['Year_Birth'] > 1980 and row['Year_Birth'] <= 1996  : # 24-39
        val = "Millenials"
    return val
```

```python
# Mengganti seluruh kolom 'Year Birth' sesuai fungsi func(row)
df['Year_Birth'] = df.apply(func, axis=1)
```

```python
# melihat unique value pada kolom 'Year Birth' setelah diupdate
df['Year_Birth'].unique()
```

```python
# mengakses dataset selain dataset yang mengandung outliers
df=df[~(df['Year_Birth'] == 'Silent Generation')]
```
